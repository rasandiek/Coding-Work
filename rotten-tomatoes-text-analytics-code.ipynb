# Loading the Dataset

When loading the dataset it's important to keep note that the dataset should have a balanced representation of the sentiment (Fresh or Rotten). So the dataset should be comprised of 50% Fresh (Positive) sentiments and 50% Rotten (Negative) sentiments.

```python
# Load the dataset
data = pd.read_csv("/kaggle/input/rotten-tomatoes-movies-and-critic-reviews-dataset/rotten_tomatoes_critic_reviews.csv")

# Filter data where the content of the review is not null
data = data[data['review_content'].notnull()]

data = data[data['review_score'].notnull()]

# Seperate fresh and rotten reviews
fresh_reviews = data[(data['review_type'] == 'Fresh') & (data['review_score'].str.match(r'\d+/\d+'))]
rotten_reviews = data[(data['review_type'] == 'Rotten')& (data['review_score'].str.match(r'\d+/\d+'))]

# Randomly Sample 75000 datapoints each to make up a 'balanced' dataset of both Fresh and Rotten at 50:50 ratio
sampled_fresh = fresh_reviews.sample(n=75000, random_state=42) 
sampled_rotten = rotten_reviews.sample(n=75000, random_state=42)

# Combine both sampled data
reviews_data = pd.concat([sampled_fresh, sampled_rotten])

#Shuffle and make the final dataset
reviews_data = reviews_data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
def find_emoticons(text):
    emoticon_count ={}
    for emoticon, description in EMOTICONS.items():
        count = len(re.findall(emoticon, text))
        if count >0:
            emoticon_count[description] = count
    return emoticon_count

reviews_data['emoticons']= reviews_data['review_content'].apply(find_emoticons)
```

### Bag of Words Model

```python
from sklearn.feature_extraction.text import CountVectorizer

# Instantiating Count Vectorizer with desired token patterns
vectorizer= CountVectorizer(token_pattern=r'\b\w+\b')

#Fit the vectorizer on the training data
train_matrix = vectorizer.fit_transform(train_data['review_content'])

#Get the feature names (words) from the vocabulary
feature_names=vectorizer.get_feature_names_out()

#Get the count of each word
word_counts = train_matrix.sum(axis=0).A1

# Create a dictionary mapping words to their counts
word_count_dict = dict(zip(feature_names, word_counts))

# Sort the dictionary by values in descending order
sorted_word_count_dict= sorted(word_count_dict.items(), key=lambda x: x[1], reverse=True)

print("Top 10 words by occurence count in Train Dataset: ")
for word, count in sorted_word_count_dict[:10]:
    print(f'{word}:{count}')
```

```python
#Fit the vectorizer on the training data
test_matrix = vectorizer.transform(test_data['review_content'])

#Get the feature names (words) from the vocabulary
feature_names=vectorizer.get_feature_names_out()

#Get the count of each word
word_counts = test_matrix.sum(axis=0).A1

# Create a dictionary mapping words to their counts
word_count_dict = dict(zip(feature_names, word_counts))

# Sort the dictionary by values in descending order
sorted_word_count_dict= sorted(word_count_dict.items(), key=lambda x: x[1], reverse=True)

print("Top 10 words by occurence count in Test Dataset: ")
for word, count in sorted_word_count_dict[:10]:
    print(f'{word}:{count}')
```

# Moeling Using Logistic Regression

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs', max_iter=6000)
```

#### Split target and Independent Variables

```python
X_train = train_matrix
X_test = test_matrix
y_train = train_data['review_type']
y_test = test_data['review_type']
```

```python
# Conduct Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(with_mean=False)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Fitting the model

```python
# Fitting the Logistic Regression Model
lr.fit(X_train,y_train)
```

### Making Predictions

```python
#Making Predictions
predictions = lr.predict(X_test)
print(predictions)
```

### Testing the Accuracy of the Model

```python
# Finding the accuracy, precision, recall:

from sklearn.metrics import confusion_matrix,classification_report
new = np.asarray(y_test)
confusion_matrix(predictions,y_test)
```
